```python
train['size_of_family'] = train['SibSp'] + train['Parch']
train
```

```python
def Family_type(number):
    if number == 0:
        return 'Alone'
    elif number > 0 and number <= 4:
        return 'Medium'
    else:
        return 'Large'
```

```python
train['Family_type'] = train['size_of_family'].apply(Family_type)
train
```

```python
train.corr(numeric_only = True)
```

```python
sns.heatmap(train.corr(numeric_only = True))
```
